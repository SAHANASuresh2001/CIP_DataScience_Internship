```python
import matplotlib.pyplot as plt

# Select numerical columns from the train dataset
numerical_columns = train.select_dtypes(include='number').columns

# Calculate the number of rows and columns for subplots
num_rows = len(numerical_columns) // 3 + (len(numerical_columns) % 3 > 0)
num_cols = min(len(numerical_columns), 3)

# Create the subplots
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))

# Iterate over the numerical columns and plot histograms
for i, column in enumerate(numerical_columns):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]
    train[column].hist(ax=ax)
    ax.set_title(column)

# Remove any unused subplots
for i in range(len(numerical_columns), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axs[row, col])

# Adjust subplot spacing
plt.tight_layout()

# Display the plot
plt.show()
```

### Data cleaning

```python
# checking duplicate values
train.duplicated().sum()
```

```python
# checking null values in each feature
train.isnull().sum()
```

```python
#to plot the outliers in the dataset
import matplotlib.pyplot as plt
train.plot(kind='box',subplots=True,layout=(5,5),figsize=(30,30))
plt.show()
```

```python
###After removing Outliers in the dataset
cols = ["fc","px_height"]

Q1 = train[cols].quantile(0.25)
Q3 = train[cols].quantile(0.75)

IQR = Q3 - Q1

train_box = train[~((train[cols] < (Q1 - 1.5 * IQR)) |(train[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
train_box.plot(kind='box',subplots=True,layout=(5,5),figsize=(30,30))
plt.show()
```

```python
# input - output split
X_train = train.loc[:,train.columns!="price_range"]
y_train =train["price_range"]
```

### feature scaling

```python
# importing StandardScaler library        
from sklearn.preprocessing import StandardScaler

# Applying the standardscaler to training data
sc = StandardScaler().fit(X_train)

# Transforming the training data into standard
X_train_std = sc.transform(X_train)
```

```python
X_train_std
```

```python
#importing LogisticRegression package

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_std, y_train)
```

```python
# Checking a train model score
model.score(X_train_std,y_train)
```
